# graphical representation of Riemann's rearrangement theorem

```python
import matplotlib.pyplot as plt
```

## plotting the harmonic series

```python
def riemann_sum(x:int):
    s = 0
    for k in range(x):
        s += (-1)**k/(k+1)
    return s
```

```python
x, y = [i for i in range(100)], [riemann_sum(i) for i in range(100)]

plt.figure(figsize=(15,15))
plt.plot(x,y)
plt.title('Harmonic series')
plt.xlabel('x points')
plt.ylabel('harmonic series sum up to x'); # the ; is used to avoid displaying line (try without)
```

## now let us rearrange the terms to converge towards 3 and plot that

```python
def pos(x):
    s, k = 0, 0
    while True:
        s += (-1)**k/(k+1)
        yield s
        k += 2
        
def neg(x):
    s, k = 0, 1
    while True:
        s += (-1)**k/(k+1)
        yield s
        k += 2
```

```python
# let us go for the first million terms of this rearranged series


def rearr_3(x):
    ''' this function puts into a list all terms of the rearranged series converging to 3,
    until diff to 3 is less than x
    '''
    pos, k, s, j = [0], 0, riemann_sum(0), 1
    for z in range(10):
        while abs(s - 3) > x:
            while pos[-1] < 3:
                pos.append(s)
                k += 2
                s += (-1)**(k)/(k+1)
            pos.append(s + (-1)**(j)/(j+1))
            j += 2

    return pos
```

```python
y =  rearr_3(0.05)
x = [i for i in range(len(y))]


plt.figure(figsize=(15,15))
plt.plot(x,y)
plt.title('Harmonic series rearranged towards converging to 3');
```

```python
s, k, j, y = 0, 0, 1, []

for t in range(100):
    while s < 2:
        s += (-1)**k/(k+1)
        k += 2
        y.append(s)
    s += (-1)**j/(j+1)
    j += 2
    y.append(s)

x = [i for i in range(len(y))]
    
plt.figure(figsize=(15,15))
plt.plot(x,y)
plt.title('Harmonic series rearranged towards converging to 3');
```

## what is the sum after the million-th term of the harmonic series?

```python
# alternative, recursive definition of the riemann_sum function above

def riemann_sum_alternative(x:int):
    return 1 if x == 0 else riemann_sum_alternative(x-1) + (-1)**x/(x+1)
```

```python
# for efficiency purposes, let us use a generator
riemann_series_x_gen = (i for i in range(1_000_000_000_000))
riemann_series_y_gen = (riemann_sum_alternative(i) for i in riemann_series_x_gen)
```

```python
#yet another way to define the generator

def gen(x):
    s, k = 0, 0
    while True:
        s += (-1)**k/(k+1)
        yield s
        k += 1
```

```python
g = gen(x)
for k in range(10**6-1):
    next(g)
print(next(g))
```

# calculations for my mini website

```python
pos = [(-1)**k/(k+1) for k in range(1000000) if k%2 == 0]
neg = [(-1)**k/(k+1) for k in range(1000000) if k%2 == 1]

new_limit = 5
s = 0
index_pos = 0
index_neg = 0
series_values = []
nb_cycles = 40

for i in range(nb_cycles): # this creates 10 "cycles"
    while s < new_limit:
        s += pos[index_pos]
        index_pos += 1
        series_values.append(s)
    s += neg[index_neg]
    index_neg += 1
    series_values.append(s)
 
```

```python
plt.figure(figsize=(15,15))
plt.plot([indice for indice in range(len(series_values))],series_values)
plt.title('Harmonic series rearranged towards converging to '+str(new_limit));
```
